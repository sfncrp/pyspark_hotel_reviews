```python
from pyspark.sql import SparkSession

spark = (SparkSession.builder
         .appName('Python spark')
         .config('spark.some.config.option','some-value')
         .getOrCreate()
         )
```

```python
df_features= spark.read.csv("hdfs://masterbig-1.itc.unipi.it:54310/user/student18/df_features.csv", header = True, inferSchema = True)
df_features.printSchema()
```

```python
# register table (if not exists)
try:
    df_features.createTempView('features')
except:
    spark.catalog.dropTempView('features')
    df_features.createTempView('features')
```

```python
import json

with open('word2vec_categories.json') as f:
     data = json.load(f)
```

```python
data['food'].extend(data['breakfast'])
del data['breakfast']
```

```python
data['food'].append('breakfast')
```

```python
data['food'].extend(['coffee', 'restaurants', 'buffet','drinks','tea'])
data['bathroom'].extend(['shower', 'bath', 'toilet', 'towels'])
data['facilities'].extend(['bar', 'pool','tv', 'furniture','lift'])
data['internet'].extend([])
data['location'].extend(['area', 'view','views','place','access', 'street','parking','transport','areas','attractions'])
data['room'].extend(['bed', 'beds','bedroom','pillows','suite','terrace'])
data['staff'].extend(['reception','receptionist','concierge', 'service', 'services'])
```

```python
for key, value in data.items():
    data[key] = list(set(value))
data
```

```python
import json

with open('final_categories.json', 'w') as outfile:
    json.dump(data, outfile)
```
